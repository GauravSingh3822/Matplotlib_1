4.5 Pie Charts



```python
car=['Gaurav','Rahul','Bhumika','Anjali','Sahil','Aman']
data=[450,385,35,20,200,500]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=car)
plt.show()
```

4.5.2 Basic Operations


```python
car=['Gaurav','Rahul','Bhumika','Anjali','Sahil','Aman']
data=[450,385,35,20,200,500]
explode=[0,0,0,.1,.2,0]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=car,explode=explode)
plt.show()
```

Adding Shadow

```python
car=['Gaurav','Rahul','Bhumika','Anjali','Sahil','Aman']
data=[650,385,35,20,200,500]
explode=[0,0,0,.1,.2,0]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=car,explode=explode,shadow=True)
plt.show()
```

Adding Percentage

```python
car=['Gaurav','Rahul','Bhumika','Anjali','Sahil','Aman']
data=[450,385,35,20,200,500]
explode=[0,0,0,.1,.2,0]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=car,explode=explode,shadow=True,autopct='%2.2f%%')
plt.show()
```

Adding wedget

```python
car=['Gaurav','Rahul','Bhumika','Anjali','Sahil','Aman']
data=[450,385,35,20,200,500]
explode=[0,0,0,.1,.2,0]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=car,explode=explode,shadow=True,autopct='%2.2f%%',wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':True})
plt.show()
```

4.6 Multiple Subplots



Index Based

```python
x=np.linspace(0,10,50)
fig,ax=plt.subplots(2,2)

y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sine Wave')

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cosine Wave')

y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')

y=np.random.rand(1000)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title('Random Distribution')

plt.tight_layout()
plt.show()
```

With Loops-Example 1

```python
plt.subplots_adjust(hspace=.4,wspace=.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')
```

With Loops_Example 2

```python
x=np.linspace(0,10,50)
y=np.sin(x)
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title("Sine Wave")
ax2.set_title('Sine Wave-Square')
ax3.set_title('Sine Wave-Cube')

plt.tight_layout()
plt.show()
```
